**GETTING THE NAMES OF THE COMPANIES**

**GETTING THE CORE INFO**

```python
import re

with open("output.csv", encoding='utf-8') as input, open('output1.csv', 'w', encoding='utf-8') as output:

    writer = csv.writer(output)
    ##writer.writerow(['location', 'price', 'description', 'company','core'])
    reader = csv.reader(input, delimiter=',')
    multi = False
    count1 = 0
    for indexx, line in enumerate(reader):
        
        thirdCol = line[2].lower()       
        # thirdCol = thirdCol.split(' ')
        #print(thirdCol)
        #print(len(line))
        gennn = re.findall("i[0-9]+", thirdCol)
        #print(gennn)
        append = None
        
        if (len(gennn) == 1):
          line.append(gennn[0])
          count1 = count1 + 1
        elif (len(gennn) > 1):
          line.append(append)
        else:
          line.append(append)
        
        writer.writerow(line)

print("Count 1: ", count1)    
```

**GETTING THE GEN INFO**

```python
import re

with open("output1.csv", encoding='utf-8') as input, open('output2.csv', 'w', encoding='utf-8') as output:

    writer = csv.writer(output)
    #writer.writerow(['location', 'price', 'description', 'company','core','gen'])
    reader = csv.reader(input, delimiter=',')
    multi = False
    count1 = 0
    for indexx, line in enumerate(reader):
        
        thirdCol = line[2].lower()       
        # thirdCol = thirdCol.split(' ')
        #print(thirdCol)
        #print(len(line))
        gennn0  = re.findall("([1-9]+th|1st)", thirdCol)
        gennn1 = [elem.replace("st", '') for elem in gennn0]
        gennn2 = [elem.replace("th", '') for elem in gennn1]
        #print(gennn)
        
        append = None
        if (len(gennn2) == 1):
          line.append(gennn2[0])
          count1 = count1 + 1
        elif (len(gennn2) > 1):
          line.append(gennn2[1])
        else:
          line.append(append)
        
        writer.writerow(line)

print("Count 1: ", count1) 
```

**GETTING THE RAM INFO**

```python
import re

with open("output2.csv", encoding='utf-8') as input, open('output3.csv', 'w', encoding='utf-8') as output:

    writer = csv.writer(output)
    #writer.writerow(['location', 'price', 'description', 'company','core','gen','ram'])
    reader = csv.reader(input, delimiter=',')
    multi = False
    count1 = 0
    for indexx, line in enumerate(reader):
        
        flag = False
        thirdCol = line[2].lower()       
        ram0 = re.findall("(\d{1,3} gb ram|\d{1,3}gb ram)", thirdCol)
        if (len(ram0) == 0):
          flag = True
          ram0 = re.findall("(\d{1,6} gb|\d{1,6}gb)", thirdCol)
        ram1 = [elem.replace(' ', '') for elem in ram0]
        ram2 = [elem.replace("gb", '') for elem in ram1]
        ram3 = [elem.replace("ram", '') for elem in ram2]
        
        append = None

        if (flag == False):
          #print(str(indexx) + "\t" + str(ram3))
          count1 = count1 + 1
          append = str(ram3[0])
        else:
          for i in ram3:
            if (4 <= int(i) and int(i) < 20):
              #print(str(indexx) + "\t" + str(ram3))
              append = str(ram3[0])
        
        line.append(append)
        
        writer.writerow(line)

print("Count 1: ", count1) 
```

**GETTING THE HARDDRIVE INFO**

```python
import re

with open("output3.csv", encoding='utf-8') as input, open('output4.csv', 'w', encoding='utf-8') as output:

    writer = csv.writer(output)
    #writer.writerow(['location', 'price', 'description', 'company','core','gen','ram','harddrive'])
    reader = csv.reader(input, delimiter=',')
    multi = False
    count1 = 0
    count2 = 0
    for indexx, line in enumerate(reader):
        
        flag = False
        thirdCol = line[2].lower()       

        ram0 = re.findall("(\d{1,6} tb|\d{1,6}tb)", thirdCol)
        ram1 = [elem.replace(' ', '') for elem in ram0]
        ram2 = ram1
        ram3 = ram1
        if (len(ram0) == 0):
          flag = True
          ram0 = re.findall("(\d{1,6} gb|\d{1,6}gb)", thirdCol)
          ram1 = [elem.replace(' ', '') for elem in ram0]
          ram2 = [elem.replace("gb", '') for elem in ram1]
        else:
          ram3 = [elem.replace("tb", '') for elem in ram1]
        
        append = None

        if (flag == False):
          
          #print(str(indexx) + "\t" + str(ram1[0]))
          append = str(ram1[0])
          count1 = count1 + 1
        else:
          for i in ram2:
            if (50 < int(i)):
              #print(str(indexx) + "\t" + str(ram3))
              append = str(i) + "gb"
              count2 = count2 + 1
        line.append(append)
        writer.writerow(line)

print("Count of HD in GB: ", count2) 
print("Count of HD in TB: ", count1) 
```
